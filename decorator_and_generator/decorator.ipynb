decorator 
A decorator is a function that adds extra functionality to another function without changing its original code

```python
#print something before and after every function.
def greet():
    print("Starting...")
    print("Hello")
    print("Ending...")
greet()
```

```python
#Decorator Syntax
def decorator(function):
    def wrapper():
        print("Starting...")
        function()
        print("Ending...")
    return wrapper
```

```python
def decorator(function):
    def wrapper():
        print("Ram RAm ")
        function()
        print("Ram Ram")
    return wrapper
```

```python
def greet():
    print("hello")
```

```python
# use of decorator
# without @
greet=decorator(greet)
greet()
```

```python
#use @
@decorator
def greet():
    print("hello")
```

```python
greet()
```

Decorator with Arguments

```python
def decorator(function):
    def wrapper(*args, **kwargs):
        print("Before")
        result = function(*args, **kwargs)
        print("After")
        return result
    return wrapper
```

```python
@decorator
def add(a, b):
    return a + b
print(add(5, 6))
```

```python
def decorator(function):
    def wrapper(*args, **kwargs):
        print("Before")
        result = function(*args, **kwargs)
        print(result)
        print("After")
    return wrapper
```

```python
@decorator
def add(a, b):
    return a + b
print(add(5, 6))
```

```python
#Multiple Decorators
def deco1(fun):
    def wrapper():
        print("One")
        fun()
    return wrapper
def deco2(fun):
    def wrapper():
        print("Two")
        fun()
    return wrapper
@deco1
@deco2
def greet():
    print("Hello")

greet()
```

```python
def repeat(n):
    def decorator(function):
        def wrapper():
            for i in range(n):
                function()
        return wrapper
    return decorator
```

```python
@repeat(3)
def greet():
    print("Hello")

greet()
```
