Generator

A generator is a special type of function that uses yield instead of return to produce values one at a time. It returns a generator object, which is an iterator.

```python
def numbers():
    yield 1
    yield 2
    yield 3

num=numbers()
print(num)
print(next(num))
print(next(num))
print(next(num))
print(next(num))
```

```python
def numbers():
    yield 1
    yield 2
    yield 3

num=numbers()
print(num)
print(next(num))
for i in num:
    print(i)
```

```python
#how yield work
def demo():
    print("A")
    yield 10
    print("B")
    yield 20
    print("C")

g = demo()
print(next(g))
print(next(g))
```

The function pauses at each yield and resumes from the same place on the next next() call

Infinite Generator

```python
def infinite():
    n = 1
    while True:
        yield n
        n += 1
g=infinite()
print(next(g))
print(next(g))
```

```python
def even():
    n=0
    while True:
        if n%2==0:
            yield n
            n+=2

g=even()
print(next(g))
print(next(g))
```

```python
#Generator Expression
g = (x * x for x in range(5))
for i in g:
    print(i)
```

```python
import sys
g = (x * x for x in range(5))
f = (x * x for x in range(500000000000))
print(sys.getsizeof(g))
print(sys.getsizeof(f))
```

chain generator
A chain generator is when one generator gets values from another generator or combines multiple generators.

```python
def gen1():
    yield 1
    yield 2
def gen2():
    yield from gen1()
    yield 3
    yield 4
g = gen2()
for i in g:
    print(i)
```

```python
#multiple chain
def even():
    yield 2
    yield 4
def odd():
    yield 1
    yield 3
def numbers():
    yield from even()
    yield from odd()
    
for i in numbers():
    print(i)
```

```python
def num1(num):
    for i in range (num):
        yield i
def square(num):
    for i in num:
        yield i*i
print(sum(square(num1(5))))
```
