#lambda function

Purpose of Lambda
.It is created for
>Small functions
>One-time use
>Functional programming

A lambda function is an anonymous (unnamed) function that consists of a single expression and automatically returns the result of that expression

>Syntax
lambda arguments : expression
Lambda Function

↓

Uses it inside map()

↓

If nothing references it later

↓

Garbage Collector removes it.


>This is why lambda is often called a Temporary Function



```python
square = lambda x: x * x#This line only creates a function and stores it in the variable square.
print(square)#printing the function object, not its result.
print(square(5))#printing the result of calling the function with argument 5
print(square(10))#printing the result of calling the function with argument 10

```

```python
square = lambda x:x*x
for i in range(1,6):
    print(square(i))
```

lambda return different data types

```python
#Integer
f = lambda x:x+1
print(f(5))
```

```python
#String
f = lambda name:"Hello " + name
print(f("Ankit"))
```

```python
#Boolean
f = lambda x:x%2==0
print(f(10))
```

```python
#list
f = lambda x:[i*i for i in range(x)]
"""
There is no loop inside the lambda.
The loop is inside a list comprehension, and a list comprehension is an expression, so it is allowed."""
print(f(5))
```

```python
#Tuple
f = lambda x:(x,x*x)
print(f(5))
```

```python
#Dictionary
f = lambda x:{"number":x,"square":x*x}
print(f(4))
```

Sort students by marks.
Sort strings by length.
Sort tuples by second element.
Find largest using lambda.
Find smallest using lambda.
Reverse every string using lambda.
Convert to uppercase.
Convert to lowercase.
Sort dictionary items by value.
Sort list of dictionaries.

```python
#Sort students by marks.
students = [
    ("Ankit", 90),
    ("Rahul", 85),
    ("Amit", 95),
    ("Neha", 88)
]
sorted_students=sorted(students, key=lambda x: x[1],reverse=True)
print(sorted_students)
```

```python
#Sort strings by length.
strings=["apple","banana","kiwi","grape"]
sorted_strings=sorted(strings,key=lambda x: len(x))
print(sorted_strings)
```

```python
#Sort tuples by second element.
tuples=[(1,3),(2,1),(4,2)]
sorted_tuples=sorted(tuples,key=lambda x:x[1])
print(sorted_tuples)
```

```python
#Find largest using lambda.
numbers = [10, 20, 5, 30, 15]
largest=(lambda x: max(x))(numbers)
print(largest)
```

```python
#Find smallest using lambda.
numbers = [10, 20, 5, 30, 15]
smallest=(lambda x: min(x))(numbers)
print(smallest)
```

```python
#Reverse every string using lambda.
strings=["apple","banana","kivi","grape"]
print((lambda x: [s[::-1] for s in x])(strings))
```

```python
#Convert to uppercase.
strings=["apple","banana","kivi","grape"]
print((lambda x: [s.upper() for s in x])(strings))
```

```python
#Convert to lowercase.
strings=['APPLE', 'BANANA', 'KIVI', 'GRAPE']
print((lambda x: [s.lower() for s in x])(strings))
```

```python
#Sort dictionary items by value.
dictionary={
    "ankit":65,
    "amit":60,
    "shan":55
}
print(sorted(dictionary.items(),key=lambda x:x[1]))
```

```python
#Sort list of dictionaries.
d=[
    {"name":"ankit","marks":65},
    {"name":"amit","marks":60},
    {"name":"shan","marks":55}
    
]
print(sorted(d,key=lambda x: x["marks"]))
```
