Iterable

An iterable is any object that can be traversed (visited one element at a time).

if we can write 
> for x in something:
then something is iterable.

map() itself is behaving like an iterator.
. map object is an iterator.

>Every iterable can be converted into an iterator.

Iterator

An iterator is an object that remembers its current position and produces the next value whenever next() is called.

Map Function


map() applies a function to every element of one or more iterables and returns a map object (an iterator).

Syntax

. One iterable
  map(function, iterable)
  eg. map(lambda x: x*x, numbers)

. Multiple iterables
  map(function, iterable1, iterable2)
  eg. map(lambda x,y:x+y,list1,list2)

. Three iterables
  map(function, iterable1, iterable2, iterable3)
  eg. map(lambda x,y,z:x+y+z,l1,l2,l3)

Return Type
< class 'map'>
map object

>map() is memory efficient.

:-next() function is used to iterate
  through the map object and get the next value.
:-Only one value exists at a time in map
  This is called Lazy Evaluation.
>Lazy evaluation means values are    generated only when they are needed, not all at once.

Memory Complexity
   O(1)  because it generates values one by one.


map is an iterator.Once exhausted It cannot restart.we must create a new map object.


Use List Comprehension when:
>we want a list immediately.
>Readability is important.
>The data is not extremely large.

Use map() when:

>Another function expects an iterator.
>we're processing very large datasets.
>we want lazy evaluation.
>we're writing in a functional programming style.

1. Create a dictionary of numbers and     their squares using map().
2. Add three lists.
3. Multiply two lists.
4. Convert Celsius to Fahrenheit.
5. Remove leading and trailing spaces from every string.

```python
#Create a dictionary of numbers and     their squares using map().
numbers=[1,2,3,4,5]
squares=map(lambda x:(x,x*x),numbers)
print(dict(squares))
```

```python
#Add three lists.
l1=[1,2,3,4]
l2=[5,6,7,8]
l3=[4,5,2,3]
result=map(lambda l1,l2,l3:l1+l2+l3,l1,l2,l3)
print(list(result))
```

```python
#Multiply two lists.
l1=[1,2,3,4]
l2=[5,6,7,8]
result=map(lambda l1,l2:l1*l2,l1,l2)
print(list(result))
```

```python
# Convert Celsius to Fahrenheit.
celsius=[0,20,30,40]
fahrenheit=map(lambda x:(x,(9/5)*x+32),celsius)
print(list(fahrenheit))
```

```python
#Remove leading and trailing spaces from every string.
strings=['   ankit   ','    amit   ','   rahul  ','   saurav   ']
result=map(lambda x:x.strip(),strings)
print(list(result))
```
